```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 

import sys



def rotation (theta):
    R = np.zeros ([2,2])
    R[0,0] = np.cos (2.0 * np.pi * theta / 180)
    R[0,1] = -np.sin (2.0 * np.pi * theta / 180)
    R[1,0] = np.sin (2.0 * np.pi * theta / 180)
    R[1,1] = np.cos (2.0 * np.pi * theta / 180)
    return R
    

if __name__ == "__main__":
    

    
    image1 = "data/Monalisa.jpg"
    image2 = "data/Hispine.jpg"
    image3 = "data/garden.png"
    
    image = image2
    
    # read image 
    image_color = cv2.imread(image, cv2.IMREAD_COLOR)
    image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
        
    image_gray  =  cv2.imread(image, cv2.IMREAD_GRAYSCALE)
 
    # resize 
    image_color = cv2.resize (image_color, (400, 600))
    image_gray  = cv2.resize (image_gray, (400, 600))
    
    
    fig, axs = plt.subplots (1,2,figsize=(12,6))
    
    axs[0].imshow (image_color)
    axs[1].imshow (image_gray,cmap='gray')
    plt.show()
    
    
```
