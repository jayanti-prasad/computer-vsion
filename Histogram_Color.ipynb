```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import sys

if __name__ == "__main__":
     
    image1 = "data/Monalisa.jpg"
    image2 = "data/Hispine.jpg"
    image3 = "data/garden.png"
    
    for image in [image1, image2, image3]:
        image_c = cv2.imread(image, cv2.IMREAD_COLOR)
        image_c = cv2.cvtColor(image_c, cv2.COLOR_BGR2RGB)
        image_c = cv2.resize (image_c, (200, 300))

        d_r = image_c[:,:,0].ravel()
        d_g = image_c[:,:,1].ravel()
        d_b = image_c[:,:,2].ravel()
    
        fig, axs = plt.subplots (1,2,figsize=(24,6))
    
        n_bins = 255
        axs[0].hist(d_r,n_bins, histtype='step', stacked=True, fill=False,color='red',label='Red')
        axs[0].hist (d_g,n_bins, histtype='step', stacked=True, fill=False,color='green',label='Green')
        axs[0].hist (d_b,n_bins, histtype='step', stacked=True, fill=False,color='blue',label='Blue')
        axs[1].imshow (image_c)
        axs[0].legend()
        fig.tight_layout()
        plt.show()

```
