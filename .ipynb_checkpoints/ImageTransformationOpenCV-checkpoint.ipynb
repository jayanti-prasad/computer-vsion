```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 


def rotation (theta, rows, cols):
    M = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),theta,1)
    return M




if __name__ == "__main__":
    

    
    image1 = "data/Monalisa.jpg"
    image2 = "data/Hispine.jpg"
    image3 = "data/garden.png"
    
    image = image2
    
    # read image 
    image_color = cv2.imread(image, cv2.IMREAD_COLOR)
    image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
        
    image_gray  =  cv2.imread(image, cv2.IMREAD_GRAYSCALE)
 
    # resize 
    image_color = cv2.resize (image_color, (400, 600))
    image_gray  = cv2.resize (image_gray, (400, 600))
    

    
    fig, axs = plt.subplots (1,2,figsize=(12,6))
    
    axs[0].imshow (image_color)
    axs[1].imshow (image_gray,cmap='gray')
    plt.show()
    
```

#  Translation
Translation is the shifting of an object's location. If you know the shift in the (x,y) direction and let it be (tx,ty), you can create the transformation matrix M as follows:



```python
    (m, n) = image_gray.shape 

    M = np.float32([[1,0,-50],[0,1,-150]])
    dst = cv2.warpAffine(image_gray,M,(n,m))
    
    
    fig, axs = plt.subplots (1,2,figsize=(12,6))
    
    axs[0].imshow (image_gray,cmap='gray')
    axs[1].imshow (dst,cmap='gray')
    plt.show()
    
    
```

# Rotation 


```python
    
    image_gray  =  cv2.imread(image, cv2.IMREAD_GRAYSCALE)
    (m, n) = image_gray.shape    

    R30 = cv2.getRotationMatrix2D(((n-1)/4.0,(m-1)/2.0),30,1)
    print("Rotation (30)\n",R30)
    
    R90 = cv2.getRotationMatrix2D(((n-1)/2.0,(m-1)/2.0),90,1)
    print("Rotation (90)\n", R90)
    
    img_30 = cv2.warpAffine(image_gray,R30,(n,m))
    img_90 = cv2.warpAffine(image_gray,R90,(n,m))
    
    fig, axs = plt.subplots (1,3,figsize=(12,6))
    axs[0].imshow (image_gray,cmap='gray')
    axs[1].imshow (img_30,cmap='gray')
    axs[2].imshow (img_90,cmap='gray')
    plt.show()
    
    
```

# Affine Transformation 

```python
    image_gray  = cv2.resize (image_gray, (400, 600))
    (m,n) = image_gray.shape 
    
    pts1 = np.float32([[500,100],[600,250],[300,350]])
    pts2 = np.float32([[500,140],[600,250],[300,350]]) 
    
    M = cv2.getAffineTransform(pts1,pts2)
    dst = cv2.warpAffine(image_gray,M,(n,m))
    
    fig, axs = plt.subplots (1,2,figsize=(12,6))
    axs[0].imshow (image_gray,cmap='gray')
    axs[1].imshow (dst,cmap='gray')

    plt.show()
```

# Perspective Transformation 

```python

    pts1 = np.float32([[50,200],[350,200],[100,600],[300,600]])
    pts2 = np.float32([[0,0],[400,0],[100,600],[400,600]])
    M = cv2.getPerspectiveTransform(pts1,pts2)

    dst = cv2.warpPerspective(image_gray,M,(400,600))
    
    fig, axs = plt.subplots (1,2,figsize=(12,9))
    axs[0].imshow (image_gray,cmap='gray')
    axs[1].imshow (dst,cmap='gray')

    plt.show()
```
