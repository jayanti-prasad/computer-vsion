```python
import sys
sys.path.append(r"c:\users\jayanti.prasad\anaconda3\lib\site-packages")
import numpy as np
import cv2 
from scipy import signal
import matplotlib.pyplot as plt


    
if __name__ == "__main__":
   
    # read image 
    image = "data/chess.png"
    image_color = cv2.imread(image, cv2.IMREAD_COLOR)
    image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
    image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
    
    
    fig, axs = plt.subplots (1,1,figsize=(12,8))
    axs.imshow (image_color)
    plt.show()
    
```

# 1. Haris corner detection

```python
    gray = cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
    gray = np.float32(gray)
    dst = cv2.cornerHarris(gray,4,3,0.08)
    
    #result is dilated for marking the corners, not important
    dst = cv2.dilate(dst,None)
    # Threshold for an optimal value, it may vary depending on the image.

    image_color [dst > 0.75*dst.max()] = [0,0,255]

    fig, axs = plt.subplots (1,2,figsize=(12,8))
    
    axs[0].imshow (image_color)
    axs[1].imshow (dst,cmap='gray')
    plt.show()
    
  
```

# 2  Good Features

```python
    corners = cv2.goodFeaturesToTrack(image_gray,25,0.5,10)

    corners = np.int0(corners)
    for i in corners:
        x,y = i.ravel()
        cv2.circle(image_color,(x,y),3,255,-1)
   

    plt.imshow(image_color)
    plt.show()
```

# 3. Shift

```python

    sift = cv2.SIFT_create()
    kp = sift.detect(image_gray,None)
    img= cv2.drawKeypoints(image_gray,kp,image_color)

    fig, axs = plt.subplots (1,2,figsize=(12,8))
    
    axs[0].imshow (image_gray,cmap='gray')
    axs[1].imshow (img,cmap='gray')
    plt.show()
    
  
```

# 4. SURF

```python
    # Initiate FAST object with default values
    fast = cv2.FastFeatureDetector_create()

    # find and draw the keypoints
    kp = fast.detect(image_gray,None)

    img2 = cv2.drawKeypoints(img, kp, None, color=(255,0,0))
    # Print all default params

    print( "Threshold: {}".format(fast.getThreshold()) )
    print( "nonmaxSuppression:{}".format(fast.getNonmaxSuppression()) )
    print( "neighborhood: {}".format(fast.getType()) )
    print( "Total Keypoints with nonmaxSuppression: {}".format(len(kp)) )

    #cv.imwrite('fast_true.png', img2)

    # Disable nonmaxSuppression
    fast.setNonmaxSuppression(0)

    kp = fast.detect(img, None)
    print( "Total Keypoints without nonmaxSuppression: {}".format(len(kp)) )
    img3 = cv2.drawKeypoints(img, kp, None, color=(255,0,0))

    fig, axs = plt.subplots (1,2,figsize=(12,8))
    axs[0].imshow (img2,cmap='gray')
    axs[1].imshow (img3,cmap='gray')
    plt.show()
    
```
