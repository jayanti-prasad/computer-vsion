```python
import numpy as np
import cv2 
from scipy import signal
import matplotlib.pyplot as plt


    
if __name__ == "__main__":
   
    # read image 
    image = "data/tiger.jpeg"
    
    
    image_color = cv2.imread(image, cv2.IMREAD_COLOR)
    image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)

    print("image shape",image_color.shape)
    
    img_blur = cv2.GaussianBlur(image_color, (3,3),0) 

    # Sobel Edge Detection
    sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge Detection on the X axis
    sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge Detection on the Y axis
    sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5) # Combined X and Y Sobel Edge Detection
 

    fig, axs = plt.subplots (2,2,figsize=(18,12))
    
    
    axs[0,0].imshow (image_color)
    axs[0,1].imshow (sobelx)
    axs[1,0].imshow (sobely)
    axs[1,1].imshow (sobelxy)
    for i in range (0,2):
        for j in range (0,2):
            axs[i,j].set_yticklabels([])
            axs[i,j].set_xticklabels([])
    plt.show()
    
    plt.show()
```
