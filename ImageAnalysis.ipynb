```python
#! pip install opencv-python
import sys
sys.path.append(r'c:\users\jayanti.prasad\anaconda3\lib\site-packages')

import numpy as np
import matplotlib.pyplot as plt
import cv2 
from PIL import Image
import sys

def read_images (image_path):

    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    image_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
 
    image_gray  = cv2.resize (image_gray, (400, 600))
    image_color  = cv2.resize (image_color, (400, 600))

    (thresh, image_bw) = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)
    return image_color, image_gray, image_bw


def image_split(image_path):

    image = cv2.imread(image_path, cv2.IMREAD_COLOR)
    image_color = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_color  = cv2.resize (image_color, (400, 600))   
  
    (m,n, d) = image_color.shape
    image_r = image_color.copy()
    image_g = image_color.copy()
    image_b = image_color.copy()

    image_r[:,:,1:] = 0
    image_g[:,:,0] = 0
    image_g[:,:,2] = 0
    image_b[:,:,0:2] = 0
    
    return  image_r, image_g, image_b


def gradiant_gray (image):
    (m,n) = image.shape
  
    dimage_by_dx = np.zeros ([m,n])
    dimage_by_dy = np.zeros ([m,n])
    for i in range (0, m):
        for j in range (1, n):
            dimage_by_dx[i,j] = image[i,j] - image[i,j-1]
    
    for i in range (0, m):
        for j in range (1, n):
            dimage_by_dy[i,j] = image[i,j] - image[i-1,j]
    return dimage_by_dx, dimage_by_dy 



if __name__ == "__main__":
    
    image1 = "data/Monalisa.jpg"
    image2 = "data/Hispine.jpg"
    image3 = "data/garden.png"

    image_color, image_gray, image_bw = read_images (image2)


    titles = ["Color","Gray","Black & White"]
    fig, axs = plt.subplots (1,3,figsize=(18,6))
    axs[0].imshow (image_color)
    axs[1].imshow (image_gray,cmap='gray')
    axs[2].imshow (image_bw,cmap='gray')
    for i in range (0,3):
        axs[i].set_title (titles [i])
    plt.show()
    
  
  

```
